```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import yfinance as yf

start = datetime.date(2018,1,1)
end = datetime.date.today()
ticker_1 = 'META'
META = yf.download(ticker_1, start, end)

META.drop(columns=["Adj Close"], inplace=True)
META['Open'].plot(label = 'META Open price', figsize = (15,7) )
META['High'].plot(label = 'META High price')
META['Low'].plot(label = 'META Low price')
META['Close'].plot(label = 'META Close price')
plt.title('META Stock Price')
plt.ylabel('Stock Price')
plt.legend()


plt.show()
```

```python
from pandas.plotting import scatter_matrix
car_comp = pd.concat([META['Open'], QQQ['Open'], SPY['Open']], axis=1)
car_comp.columns = ['META Open', 'QQQ Open', 'SPY Open']
```

```python
META['Returns'] = (META['Close']/META['Close'].shift(1))-1
QQQ['Returns'] = (QQQ['Close']/QQQ['Close'].shift(1))-1
SPY['Returns'] = (SPY['Close']/SPY['Close'].shift(1))-1
```

```python
META.head()
```

```python
QQQ.head()
```

```python
SPY.head()
```

```python
META['Returns'].hist(bins = 100, label = 'META', alpha = 0.5, figsize = (15, 7))
QQQ['Returns'].hist(bins = 100, label = 'QQQ', alpha = 0.5)
SPY['Returns'].hist(bins = 100, label = 'SPY', alpha = 0.5)
plt.legend()
```

```python
META['Cumulative Return'] = (1+META['Returns']).cumprod()
QQQ['Cumulative Return'] = (1+QQQ['Returns']).cumprod()
SPY['Cumulative Return'] = (1+SPY['Returns']).cumprod()
```

```python
META['Cumulative Return'].plot(label='META', figsize=(15, 7))
QQQ['Cumulative Return'].plot(label='QQQ')
SPY['Cumulative Return'].plot(label='SPY')
plt.title('Cumulative Return vs Time')
plt.legend()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "Suicide rate.csv"
suicide_data = pd.read_csv(file_path)

# Filter data for the United States and the desired period
usa_data = suicide_data[(suicide_data["Location"] == "United States of America")& (suicide_data["Dim1"] == "Both sexes")]

```

```python
usa_data
```

```python
suicide_df = pd.DataFrame({'Year': usa_data["Period"], 'Suicide Rate': usa_data["FactValueNumeric"]})
suicide_df
```

```python
plt.figure(figsize=(15, 6))
plt.plot(suicide_df["Year"],suicide_df["Suicide Rate"], marker='o', linestyle='-')
plt.title('Suicide Rates in the United States (2000-2019)')
plt.xlabel('Year')
plt.ylabel('Suicide Rate (per 100,000 population)')
plt.grid(True)
plt.xlim(2012, 2019)
plt.show()
```
